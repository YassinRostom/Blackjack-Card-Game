```python
from random import shuffle
from IPython.display import clear_output
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card():
    '''
    Card Class: Generates an instence of a Card Object (example: Herats of 2)
    Arrguments:
        suit : Card Type (Heart or Diamonds or etc)
        ranks: Card rank
    Methods:
        1- return the card value (useful for comparison)
        2- print the card object representation
    '''
    
    def __init__(self,suit,rank):
        
        self.suit = suit
        self.rank = rank
        self.values = values[f"{self.rank}"]   # Note: The use of F-String litteral with dictionary
        
    def __str__(self):
        return(f"{self.suit} of {self.rank}")
```

```python
class Deck():
    '''
    Deck Class: Genrates a new deck of 52 card (object) instences. (For BJ: I will multiply the deck by 6)
    No Arrguments:
    
    Methods:
        1- shuffle: shuffle the list of cards.
        2- deal: pop the 1st object in the list.   
    '''
    
    def __init__(self):
        self.all_card = []
        self.discard_all_card = []
        self.suits = suits
        self.ranks = ranks
        
        # Generate a deck of 52 cards
        for i in self.suits:
            for n in self.ranks:
                self.all_card.append(Card(i,n))
        
        self.all_card = self.all_card * 6                 # use 6 deck of cards for the game

        
    def shuffle(self):
        return shuffle(self.all_card)
    
    def deal(self):
        return self.all_card.pop(0)                       # pop card from the top of the list
```

```python
class Player():
    '''
    Player Class: Generates an empty list for each player
    Arguments:
        Name: Player Name
        Budget: Starting Budget
    
    Method:
        1- Print player name and starting budget.
        2- setup_Game: set the player name and starting Budget
        3- ask: asks the player next move (Hit or Stay)
        4- hit: Adds new cards to the player hand (list)
        5- total: sums the card values the player have in hand (Ace by default 11)
        6- discard: clear the player hand
    '''
    
    def __init__(self,name="",budget=0):
        self.all_card = []
        self.name = name
        self.budget = budget
        self.sum = 0                           # use in total method
        self.move = ""                         # use in ask method
        
    def __str__(self):
        return(f"Player: {self.name} has {self.budget} £")
            
    def setup_Game(self):
        self.name = input("Please Entre Player Name? ")
        while True:
            try:
                self.budget = int(input("Starting Budget? "))
            except:
                clear_output()
                print("Error!! Entre an Integer")
            else:
                clear_output()
                print(f"Budget Accepted --> {self.name} has {self.budget}£ in bank")
                break
    
    def ask(self):
        while True:
            self.move = input(f"{self.name} do you want to Hit or Stay?[H or S]")
            
            if self.move in ["S","H"]:
                return self.move                 # return statment will break the loop
            else:
                print("Error!! Invalid Answer")
    
    def hit(self,new_card):                      # Hit a card from the deck and add it to the player hand (list)
        if type(new_card) == type([]):
            self.all_card.extend(new_card)
        else:
            self.all_card.append(new_card)
    
    def total(self):
        self.sum = 0 # Reset on each iteration
        for i in range(len(self.all_card)):
            self.sum += self.all_card[i].values
        return self.sum
    
    def discard(self):                           # discard the round cards 
        self.all_card.clear()
```

```python
class Bank():
    '''
    Bank Class: Represents the Money the player has in his bank
    Arguments:
        budget: Starting budget the player wants to set
        
    Method
        1- print the current budget in the bank
        2- win_add: double the bet and add to budget (Win)
        3- subtract: subtracts the bet value from the budget (Loss)
        4- bet_check: request the player to entre the bet at the start of each round and
                      do a number of checks (Read func docstring)
    '''
    
    
    def __init__(self,budget):
        self.budget = budget
        self.bet = 0
        
    def __str__(self):
        return (f"Budget = {self.budget}")
    
    def win_add(self):
        self.budget +=  (self.bet * 2) 
        
    def subtract(self):
        self.budget -= self.bet
    
    def bet_check(self):
        '''
        bet_check method checks if the input
            1- integer
            2- if the player have sufficient funds to place this bet
        '''
        while True:

            try:
                self.bet = int(input("Can you please insert a bet? "))
            except:
                clear_output()
                print("Error!! Insert an Integer")
            else:
                if self.bet <= self.budget:
                    clear_output()
                    print(f"Bet ({self.bet}) Accepted --> Current Budget = {self.budget}")
                    break
                else:
                    print("insufficient funds\n"
                         f"Budget = {self.budget}")
                    continue
```
